```python
import os
import pandas as pd
```

```python
data_confirmed = pd.read_excel('E:/1.DATA_SCIENCE/Projects/covid_19_dataset/confirmed_covid_realtime.xlsx')

data_deaths = pd.read_excel('E:/1.DATA_SCIENCE/Projects/covid_19_dataset/death_covid_realtime.xlsx')

data_recovered =pd.read_excel('E:/1.DATA_SCIENCE/Projects/covid_19_dataset/death_covid_realtime.xlsx') 

print("Shape of confirmed dataset is: ", data_confirmed.shape)
print("Shape of deaths dataset is: ", data_deaths.shape)
print("Shape of recovered dataset is: ", data_recovered.shape)
```

```python
# Unpivoting Data
```

```python
transpose_confirmed = pd.melt(data_confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])
transpose_deaths = pd.melt(data_deaths, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])
transpose_recovered=pd.melt(data_recovered, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])

print("shape of transposed confirmed dataset is: ",transpose_confirmed.shape)
print("shape of transposed deaths dataset is: ",transpose_deaths.shape)
print("shape of transposed recovered dataset is: ",transpose_recovered.shape)

transpose_confirmed
```

```python
# Changing the "value" name to the corresponding names as death,recovered and confirmed.

transpose_confirmed.columns = transpose_confirmed.columns.str.replace('value', 'Confirmed')

transpose_deaths.columns = transpose_deaths.columns.str.replace('value', 'Deaths')

transpose_recovered.columns = transpose_recovered.columns.str.replace('value', 'Recovered')
```

```python
print(transpose_confirmed.isnull().sum())


print(transpose_recovered.isnull().sum())


print(transpose_deaths.isnull().sum())
```

```python
# Handling null values
```

```python
# Removing the Province/State column 

x1 =transpose_confirmed.drop(['Province/State'], axis = 1)
x2 =transpose_recovered.drop(['Province/State'], axis = 1)
x3 =transpose_deaths.drop(['Province/State'], axis = 1)

x1
```

```python
# Joining the datasets
```

```python
full_join1 = x1.merge(x3[['Country/Region','Date','Deaths']], 
                                      how = 'left', 
                                      left_on = ['Country/Region','Date'], 
                                      right_on = ['Country/Region','Date'])
full_join1
```

```python

full_join1 = full_join1.merge(x2[['Country/Region','Date','Recovered']], 
                                      how = 'left', 
                                      left_on = ['Country/Region','Date'], 
                                      right_on = ['Country/Region','Date'])
```

```python
full_join1
```

```python
#Changing the data type of date column

full_join1['Date']= pd.to_datetime(full_join1['Date'])

```
